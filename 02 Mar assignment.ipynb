## Question 01 - What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

# Answer :- 

Matplotlib is a popular data visualization library in Python that enables you to create a variety of 2D plots and graphs to represent your data visually. It provides a wide range of tools and functions for creating different types of plots, customizing their appearance, and adding annotations and labels.

Matplotlib is used for various data analysis and scientific computing tasks, including data exploration, statistical analysis, machine learning, and more. Its versatility and flexibility make it a go-to choice for data scientists and analysts.

The Pyplot module of Matplotlib provides a convenient interface for creating various types of plots. Here are five examples of plots that can be plotted using Pyplot:

1. Line Plot: A line plot is a basic type of plot that represents data as a series of points connected by straight lines. It is used to visualize the trend or pattern in data over time or across different variables.

2. Scatter Plot: A scatter plot is a type of plot that shows the relationship between two variables by plotting their values as points on a graph. It is useful for identifying patterns or trends in data that may not be visible in a line plot.

3. Bar Plot: A bar plot is a type of plot that represents data using rectangular bars. It is used to compare the values of different variables or categories.

4. Histogram: A histogram is a type of plot that represents the distribution of data by dividing it into intervals and plotting the number of observations in each interval as a bar. It is used to visualize the shape and spread of the data.

5. Heatmap: A heatmap is a type of plot that represents the values of a matrix as colors in a grid. It is used to visualize the patterns and relationships in large datasets.

# Question 02 - What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

# Answer :- 

A scatter plot is a type of plot that shows the relationship between two variables by plotting their values as points on a graph. It is useful for identifying patterns or trends in data that may not be visible in a line plot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the seed for reproducibility
np.random.seed(3)

# Generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Display the plot
plt.show()
```

This will generate a scatter plot with x-values on the horizontal axis and y-values on the vertical axis, with each point representing a pair of x and y values. The title, xlabel, and ylabel have also been added to the plot.

## Question 03 - Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

# Answer :-

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange plots in a grid of rows and columns, and specify which plot to use for each set of data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the four line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Create a figure with a 2x2 grid of plots
plt.figure(figsize=(8, 8))

# Plot the four lines in the grid
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

# Add a common x-label and y-label for the entire figure
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Display the plot
plt.show()
```

This will create a 2x2 grid of line plots, with each plot showing a different set of data. The title() function is used to add a title to each plot, and the xlabel() and ylabel() functions are used to add a common x-label and y-label for the entire figure.

## Question 04 - What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Answer :- 

A bar plot is a chart that represents categorical data with rectangular bars, with the length of each bar proportional to the value it represents. Bar plots are commonly used to compare data across different categories.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.bar(company, profit)

# Add a title and labels for the x and y axis
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit ($ in thousands)")

# Display the plot
plt.show()

#This will create a vertical bar plot where the height of each bar represents the profit of each company.
```

```python
# Create a horizontal bar plot
plt.barh(company, profit)

# Add a title and labels for the x and y axis
plt.title("Company Profit")
plt.xlabel("Profit ($ in thousands)")
plt.ylabel("Company")

# Display the plot
plt.show()
#This will create a horizontal bar plot where the length of each bar represents the profit of each company.
```

## Question 05 - What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Answer :- 

A box plot is a chart that shows the distribution of a dataset by displaying the five-number summary (minimum, first quartile, median, third quartile, and maximum). Box plots are useful for identifying outliers, skewness, and the range of the data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the two datasets into a single list
data = [box1, box2]

# Create the box plot
plt.boxplot(data)

# Add a title and labels for the x and y axis
plt.title("Box Plot")
plt.xlabel("Data")
plt.ylabel("Value")

# Display the plot
plt.show()
```

This will create a box plot with two boxes representing the box1 and box2 datasets. The horizontal line inside each box represents the median, while the box itself spans the interquartile range (IQR). The whiskers extend to the minimum and maximum values within 1.5 times the IQR, and any points beyond the whiskers are considered outliers.
